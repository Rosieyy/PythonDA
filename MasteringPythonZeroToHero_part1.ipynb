```python
# Function(variable number of input arguments)
def add(*args):
    for i in range(len(args)):
        print(args[i])
```

```python
add(1,2,3)
```

```python
# Function(variable number of input arguments)
# double star means you are receiving a key value pair list
def keyvaluepair(**args):
    for x in args:
        print(x,args[x])
```

```python
keyvaluepair(a=1,b=2,c="jk")
```

```python
# Function(default value)
# default values are assigned at the time when function is created not the time when function is called
def f(su=0):
    print(su)
```

```python
f()
```

```python
f(3)
```

```python
def ff(L=[1,2]):
    for i in L:
        print(i)
```

```python
ff()
```

```python
ff(L=[1,2,3])
```

```python
ff()
```
